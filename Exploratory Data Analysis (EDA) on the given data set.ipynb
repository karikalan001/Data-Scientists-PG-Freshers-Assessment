# Perform Exploratory Data Analysis (EDA) on the given data set to say the story that the data contains

### libraries to import data


```python
#libraries to import data
import pandas as pd
import numpy as np
```

### Import the given data

```python
data = pd.read_csv('hepatitis_csv.csv')
data.head()
```

### Description of a given dataset

```python
data.info()
data.describe()
```

#### From the above table,The patients are in the age from 7 to 78. 
#### sgot (serum glutamic-oxaloacetic transaminase) have the greatest standard deviation
#### Above information shows that the attributes like sex, age, antiviral, histology and class have no null values.

### Converting the object to float type by assigining 'False' as 0 and 'True' as 1

```python
# Changing the type
data["steroid"].replace((False,True),(0,1),inplace=True)
data["antivirals"].replace((False,True),(0,1),inplace=True)
data["fatigue"].replace((False,True),(0,1),inplace=True)
data["malaise"].replace((False,True),(0,1),inplace=True)
data["anorexia"].replace((False,True),(0,1),inplace=True)
data["liver_big"].replace((False,True),(0,1),inplace=True)
data["liver_firm"].replace((False,True),(0,1),inplace=True)
data["spleen_palpable"].replace((False,True),(0,1),inplace=True)
data["spiders"].replace((False,True),(0,1),inplace=True)
data["ascites"].replace((False,True),(0,1),inplace=True)
data["varices"].replace((False,True),(0,1),inplace=True)
data["histology"].replace((False,True),(0,1),inplace=True)



data.head()
```

### Same process done here, In class we have the variables like die and live, so we are assigining them , die as 0 and live as 1 and in sex attribute male and female are converted to 1 and 0 respectively

```python
data["class"].replace(('die','live'),(0,1),inplace=True)
data["sex"].replace(('female','male'),(0,1),inplace=True)
```

```python
data.head()
```

### Checking For Null Values in the attributes

```python
data.isna().sum()
```

### filling the missing values in the attributes that have unique value like 0 and 1

```python
# filling the missing values
data['steroid'].replace(to_replace=np.nan,value=1,inplace=True)
data['fatigue'].replace(to_replace=np.nan,value=0,inplace=True)
data['malaise'].replace(to_replace=np.nan,value=1,inplace=True)
data['anorexia'].replace(to_replace=np.nan,value=1,inplace=True)
data['liver_big'].replace(to_replace=np.nan,value=1,inplace=True)
data['liver_firm'].replace(to_replace=np.nan,value=1,inplace=True)
data['spleen_palpable'].replace(to_replace=np.nan,value=1,inplace=True)
data['spiders'].replace(to_replace=np.nan,value=1,inplace=True)
data['ascites'].replace(to_replace=np.nan,value=1,inplace=True)
data['varices'].replace(to_replace=np.nan,value=1,inplace=True)


data.head()
```

#### Filling the null values in the attributes with more number of unique values with the median of the respective attributes

```python
# bilirubin missing data is filled with median
data['bilirubin'].skew(axis=0,skipna = True)
data['bilirubin'].median()
data['bilirubin'].replace(to_replace=np.nan,value=1,inplace=True)

# alk_phosphate missing data is filled with median
data['alk_phosphate'].skew(axis=0,skipna = True)
data['alk_phosphate'].median()
data['alk_phosphate'].replace(to_replace=np.nan,value=85,inplace=True)

# sgot missing data is filled with median
data['sgot'].skew(axis=0,skipna = True)
data['sgot'].median()
data['sgot'].replace(to_replace=np.nan,value=58,inplace=True)

# albumin missing data is filled with median
data['albumin'].skew(axis=0,skipna = True)
data['albumin'].median()
data['albumin'].mean()
data['albumin'].replace(to_replace=np.nan,value=4,inplace=True)

# protime missing data is filled with median
data['protime'].skew(axis=0,skipna = True)
data['protime'].median()
data['protime'].mean()
data['protime'].replace(to_replace=np.nan,value=61,inplace=True)

data.head()
```

### plotting the sex and class attribute  

```python
import matplotlib.pyplot as plt

male =len(data[data['sex'] == 1])
female = len(data[data['sex']==0])

plt.figure(figsize=(8,6))

# Sex column plot
labels = 'Male','Female'
sizes = [male,female]
colors = ['red', 'lightblue']
explode = (0.1, 0) 
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
 
plt.axis('equal')
plt.show()
```

#### its observed that 89.7 % are female and only 10.3 % are male.
#### Its observed that men diagonised with hepatitis tend to survive more when compared with female patients.

```python
die =len(data[data['class'] == 0])
live = len(data[data['class']== 1])
plt.figure(figsize=(8,6))
# class column plot
labels = 'DEAD','ALIVE'
sizes = [die,live]
colors = ['orange', 'lightyellow']
explode = (0.1, 0)
 
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=90)
 
plt.axis('equal')
plt.show()
```

#### its observed that 79.4% are Alive and 20.6 % are dead.

## Heatmap
#### A heat map (or heatmap) is a data visualization technique that shows magnitude of a phenomenon as color in two dimensions. The variation in color may be by hue or intensity, giving obvious visual cues to the reader about how the phenomenon is clustered or varies over space.

```python
plt.figure(figsize=(15,12))
sns.heatmap(data.corr(), cmap='plasma',linewidths=.1,annot = True)
plt.show()
```

#### Thus, it is visible from the heatmap that some of the variables show a coefficient of ~0.6 or -0.4, but most of them display a very low correlation coefficient. So we can able to know that there is no strong linear correlation between our variables.

### pairplot
#### pairplot is used to plot multiple pairwise bivariate distributions in a dataset.

```python
numerical_variables = ['age', 'bilirubin', 'protime', 'albumin', 'alk_phosphate', 'sgot']
data[numerical_variables] = data[numerical_variables].astype(float)
numerical_class = ['age', 'bilirubin', 'protime', 'albumin', 'alk_phosphate', 'sgot','class']


pairplot = sns.pairplot(data[numerical_class] ,hue="class",palette='rocket')
plt.suptitle('Bivariate Analysis of Numerical Data',fontsize=25);
plt.show(pairplot)
```

#### From the above pairplot :
#### 'died people' distplot has been right skewed in all the plots and 'live' has been normal only in albumin. And linear relationship has not been found in these variables

```python
graph = sns.PairGrid(data,
                 x_vars=["anorexia","varices", "ascites","spleen_palpable"],
                 y_vars=['bilirubin', 'albumin', 'alk_phosphate', 'sgot', 'protime'],
                 hue = 'class')
graph.map(sns.kdeplot, s = 6,color='violet')
graph.add_legend(frameon=True, bbox_to_anchor=(0.33, 0.96));
```
